#### Treating Outliers manually

```python
# By setting boundaries we are removing outliers
print('Shape of data before treating outliers:', data.shape)

data_df = data[
    (data['age'].between(20, 70)) & 
    (data['duration'] < 1200) & 
    (data['campaign'] < 12)
]
lower, upper = np.percentile(data['balance'], [1, 99])
data_df['balance'] = data['balance'].clip(lower, upper)

print('Shape of data after treating outliers:', data_df.shape)
```

#### Feature Engineering:

```python
#using binning Age and balance
data_df['age_group'] = pd.cut(data_df['age'], 
                           bins=[0, 30, 50, 70], 
                           labels=['Young (<30)', 'Middle-aged (30-50)', 'Senior (50+)'])
data_df['balance_group'] = pd.cut(data_df['balance'],
                               bins=[-np.inf, 0, 1000, 5000, np.inf],
                               labels=['Negative', 'Low (0-1k)', 'Medium (1k-5k)', 'High (5k+)'])
#Adding New columns
data_df['total_contact_time'] = data_df['duration'] * data_df['campaign']
data_df['total_contact_time'] = data_df['total_contact_time'].replace([np.inf, -np.inf], np.nan)  # If campaign=inf
data_df['total_contact_time'].fillna(0, inplace=True)  # Replace NaNs with 0

data_df['contact_intensity'] = data_df['campaign'] / data['day']
data_df['contact_intensity'].replace([np.inf, -np.inf], np.nan, inplace=True)
data_df['contact_intensity'].fillna(0, inplace=True)

data_df['call_duration_group'] = pd.cut(
    data_df['duration'],
    bins=[0, 60, 180, 300, np.inf],
    labels=['Short (<1min)', 'Medium (1-3min)', 'Long (3-5min)', 'Very Long (>5min)'])
```

### ðŸ› ï¸ Feature Engineering: New Binned and Derived Features

To enrich the dataset and improve model interpretability, several new features were engineered from existing ones:

#### ðŸ”¢ Binned Features
- **`age_group`**: Categorizes customers into age-based groups:
  - `Young (<30)` â€“ Ages below 30
  - `Middle-aged (30-50)` â€“ Ages between 30 and 50
  - `Senior (50+)` â€“ Ages above 50  
  This helps capture age-based behavioral patterns in a simplified form.

- **`balance_group`**: Groups the customer's bank balance into financial segments:
  - `Negative` â€“ Balance below 0 (debt/overdraft)
  - `Low (0-1k)` â€“ Between â‚¬0 and â‚¬1,000
  - `Medium (1k-5k)` â€“ Between â‚¬1,000 and â‚¬5,000
  - `High (5k+)` â€“ Over â‚¬5,000

  - `Short (<1min)`

### ðŸ” Project Strategy: Two-Phase Modeling Approach

To solve the problem effectively, we designed a two-phase modeling strategy focused on improving customer targeting and call efficiency during a term deposit marketing campaign:

#### 1ï¸âƒ£ **Before Making Calls: Pre-call Filtering**
- **Objective:** Identify and filter out customers who are unlikely to subscribe **before** making any contact.
- **Goal:** Optimize call center resources by focusing only on high-potential leads.
- **Features Used:** 
  - Customer demographics (age, job, marital status, education, etc.)
  - Financial history (balance, loans, default status)
  
This phase helps reduce unnecessary calls and improves the efficiency of the campaign.

#### 2ï¸âƒ£ **After Making Calls: Post-call Prediction**
- **Objective:** Predict which customers are likely to convert **after** being contacted.
- **Goal:** Determine whether to follow up with a customer or not, based on their response and call metadata.
- **Features Used:**
  - Call metadata: call duration, contact method, campaign (number of contacts), day/month of contact
  - Demographic and financial information

This phase enhances targeting for follow-ups and maximizes the return on each interaction.

Using raw features and predicting with Lazy predictor , just to see which model shows the best recall for 0 , so that we can avoid the resource allocation in making calls.

```python
data_before_call = data_df[['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'y']]
data_before_call.shape
```

```python
# Lets balance the dataset
x =data_before_call.drop('y', axis=1)
y = data_before_call['y']

le = LabelEncoder()
y_preprocessed = le.fit_transform(y)

# Split the data
x_train, x_test, y_train, y_test = train_test_split( x, y_preprocessed, test_size=0.3, random_state=seed, stratify=y_preprocessed)

# Define categorical and numerical columns
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan' ]
num_cols = ['age', 'balance']

# Define preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num_age', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', MinMaxScaler())
        ]), ['age']),
        ('num_balance', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', RobustScaler())
        ]), ['balance']),
        
        ('cat', Pipeline([
            ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
            ('encoder', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'))
        ]), cat_cols)
    ])

# Apply preprocessing
x_train_preprocessed = preprocessor.fit_transform(x_train)
x_test_preprocessed= preprocessor.transform(x_test)
```

```python
feature_names = []
for name, trans, cols in preprocessor.transformers_:
    if hasattr(trans, 'get_feature_names_out'):
        feature_names.extend(trans.get_feature_names_out(cols))
    else:
        feature_names.extend(cols)

# Convert to DataFrame
x_train_preprocessed_df = pd.DataFrame(x_train_preprocessed, columns=feature_names)
print(x_train_preprocessed_df.columns.tolist() )
```

```python
# Applying SMOTE Tomek on imbalanced dataset to balance between 0s and 1s

smotetomek = SMOTETomek( random_state = seed)
x_train_res, y_train_res = smotetomek.fit_resample(x_train_preprocessed, y_train)
```

#### Fitting preprocessed data on different models to observe which produces good recall and precision

```python
# Initialize models with imbalance handling
models = {
    "Dummy (Stratified)": DummyClassifier(strategy="stratified"),
    "XGBoost": XGBClassifier(scale_pos_weight=len(y_train_res[y_train_res==0])/len(y_train_res[y_train_res==1])),
    "LightGBM": LGBMClassifier(class_weight='balanced'),
    "Bagging (DT)": BaggingClassifier(
        estimator=DecisionTreeClassifier(class_weight='balanced'),
        n_estimators=50
    ),
    "Decision Tree": DecisionTreeClassifier(class_weight='balanced'),
    "Random Forest": RandomForestClassifier(class_weight='balanced', n_jobs=-1),
    "KNN": KNeighborsClassifier(),
    "AdaBoost": AdaBoostClassifier(),

}

# Cross-validation setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)

# Train, evaluate, and cross-validate
results = {}
for name, model in models.items():
    # Cross-validation scores
    cv_scores = cross_val_score(
        model, x_train_res, y_train_res, 
        cv=cv, scoring='recall', n_jobs=-1
    )
    
    # Train and test evaluation
    model.fit(x_train_res, y_train_res)
    y_pred = model.predict(x_test_preprocessed)
    y_prob = model.predict_proba(x_test_preprocessed)[:, 1] if hasattr(model, "predict_proba") else None

    # Calculate metrics
    cm = confusion_matrix(y_test, y_pred)
    recall_0 = recall_score(y_test, y_pred, pos_label=0)
    recall_1 = recall_score(y_test, y_pred, pos_label=1)
    
    # Store results
    results[name] = {
        "CV F1 Mean": np.mean(cv_scores),
        "CV F1 Std": np.std(cv_scores),
        "Test F1": f1_score(y_test, y_pred, average='weighted'),
        "Test ROC AUC": roc_auc_score(y_test, y_prob) if y_prob is not None else np.nan,
        "Classification Report": classification_report(y_test, y_pred),
        "Recall (Class 0)": recall_0,
        "Recall (Class 1)": recall_1,
        "Confusion Matrix": cm
    }
results_df = pd.DataFrame.from_dict(results, orient='index')

# columns we want to display (as a list)
columns_to_display = [
    "CV F1 Mean",
    "Test F1",
    "Test ROC AUC",
    "Recall (Class 0)",
    "Recall (Class 1)",
    "Confusion Matrix"
]


#  Select and sort the results
final_results = results_df[columns_to_display].sort_values("Test F1", ascending=False)

# Set display options and show results
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_colwidth', None)  # Show full content
pd.set_option('display.expand_frame_repr', False)  

#  Display the final results
print("\nModel Evaluation Results:")
display(final_results)
```

#### Finding Optimal threshold and model suitable for best trade off between Precision, recall and thus chosing a high F1

```python
def threshold_search(y_true, y_proba, model_name, thresholds=np.arange(0.1, 0.9, 0.1)):
    results = []
    for thresh in thresholds:
        preds = (y_proba >= thresh).astype(int)
        precision = precision_score(y_true, preds)
        recall = recall_score(y_true, preds)
        f1 = f1_score(y_true, preds)
        cm = confusion_matrix(y_true, preds)
        tn, fp, fn, tp = cm.ravel()
        results.append({
            "Model": model_name,
            "Threshold": thresh,
            "Precision": precision,
            "Recall": recall,
            "F1": f1,
            "TP": tp,
            "FN": fn,
            "TP>FN": tp > fn
        })

    df_results = pd.DataFrame(results)
    print(f"\nThreshold results for {model_name}:")
    print(df_results[["Model", "Threshold", "Precision", "Recall", "F1"]].to_string(index=False))
    
    return df_results
```

```python
# Define models
lgbm_model = LGBMClassifier()
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
ada_model = AdaBoostClassifier()

# Fit models
lgbm_model.fit(x_train_res, y_train_res)
xgb_model.fit(x_train_res, y_train_res)
ada_model.fit(x_train_res, y_train_res)

# Define your feature names
feature_names = ['age', 'balance', 'job_blue-collar', 'job_entrepreneur', 'job_housemaid',
                 'job_management', 'job_retired', 'job_self-employed', 'job_services', 'job_student',
                 'job_technician', 'job_unemployed', 'job_unknown', 'marital_married', 'marital_single',
                 'education_secondary', 'education_tertiary', 'education_unknown', 'default_yes',
                 'housing_yes', 'loan_yes']

# Plot LightGBM feature importance
def plot_lgbm_importance(model, feature_names, top_n=15):
    importances = model.feature_importances_
    indices = importances.argsort()[::-1][:top_n]
    plt.figure(figsize=(8, 5))
    plt.barh(range(top_n), importances[indices][::-1])
    plt.yticks(range(top_n), [feature_names[i] for i in indices][::-1])
    plt.xlabel("Importance")
    plt.title("LightGBM Feature Importance")
    plt.tight_layout()
    plt.show()

# Plot XGBoost feature importance
def plot_xgb_importance(model, feature_names, top_n=15):
    booster = model.get_booster()
    scores = booster.get_score(importance_type='gain')
    sorted_feats = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:top_n]
    
    feat_names = [feature_names[int(f[1:])] for f, _ in sorted_feats]
    feat_scores = [s for _, s in sorted_feats]
    
    plt.figure(figsize=(8, 5))
    plt.barh(range(top_n), feat_scores[::-1])
    plt.yticks(range(top_n), feat_names[::-1])
    plt.xlabel("Importance (Gain)")
    plt.title("XGBoost Feature Importance")
    plt.tight_layout()
    plt.show()
def plot_adaboost_importance(model, feature_names, top_n=15):
    importances = model.feature_importances_
    indices = importances.argsort()[::-1][:top_n]
    
    plt.figure(figsize=(8, 5))
    plt.barh(range(top_n), importances[indices][::-1])
    plt.yticks(range(top_n), [feature_names[i] for i in indices][::-1])
    plt.xlabel("Importance")
    plt.title("AdaBoost Feature Importance")
    plt.tight_layout()
    plt.show()

# Call the functions
plot_lgbm_importance(lgbm_model, feature_names)
plot_xgb_importance(xgb_model, feature_names)
plot_adaboost_importance(ada_model, feature_names)
```

#### Finding Optimal threshold and model suitable for best trade off between Precision, recall and thus chosing a high F1

```python
def threshold_search(y_true, y_proba, model_name, thresholds=np.arange(0.1, 0.9, 0.1)):
    results = []
    for thresh in thresholds:
        preds = (y_proba >= thresh).astype(int)
        precision = precision_score(y_true, preds)
        recall = recall_score(y_true, preds)
        f1 = f1_score(y_true, preds)
        cm = confusion_matrix(y_true, preds)
        tn, fp, fn, tp = cm.ravel()
        results.append({
            "Model": model_name,
            "Threshold": thresh,
            "Precision": precision,
            "Recall": recall,
            "F1": f1,
            "TP": tp,
            "FN": fn,
            "TP>FN": tp > fn
        })

    df_results = pd.DataFrame(results)
    print(f"\nThreshold results for {model_name}:")
    print(df_results[["Model", "Threshold", "Precision", "Recall", "F1"]].to_string(index=False))
    
    return df_results
```

```python
xgb_model = XGBClassifier(scale_pos_weight=len(y_train_enn[y_train_enn==0])/len(y_train_enn[y_train_enn==1]))
lgbm_model = LGBMClassifier(class_weight='balanced')
ada_model = AdaBoostClassifier()


xgb_model.fit(x_train_enn, y_train_enn)
lgbm_model.fit(x_train_enn, y_train_enn)
ada_model.fit(x_train_enn, y_train_enn)

# Get predicted probabilities (positive class probability) from each model
xgb_probs = xgb_model.predict_proba(x_test_preprocessed)[:, 1]
lgbm_probs = lgbm_model.predict_proba(x_test_preprocessed)[:, 1]
ada_probs = ada_model.predict_proba(x_test_preprocessed)[:, 1]

df_xgb = threshold_search(y_test, xgb_probs, model_name='XGBoost')
df_lgbm = threshold_search(y_test, lgbm_probs, model_name='LightGBM')
df_ada = threshold_search(y_test, ada_probs, model_name='AdaBoost')

# Combine all results
all_threshold_results = pd.concat([df_xgb, df_lgbm, df_ada], ignore_index=True)
```

The threshold tuning results for three models â€” XGBoost, LightGBM, and AdaBoost â€” by showing how precision, recall, and F1 score change at different classification thresholds.
- âœ… XGBoost: Best threshold range: 0.3 to 0.5. At 0.3, it strike a good balance: moderate precision and still fairly high recall (F1 = 0.455).
            At 0.5, precision increases but recall drops.
- âœ… LightGBM: Stable performance across thresholds from 0.3 to 0.5. F1 peaks at 0.5 (0.471) with a decent tradeoff between precision and recall.
- âš ï¸ AdaBoost:Extremely low precision at low thresholds: predicts everyone as positive â†’ high recall but meaningless results. Fails completely at 0.6               and above (predicts no positives). 
- ðŸ“Œ Final Notes : If avoiding false positives is more important ( avoiding calling unlikely subscribers), go with a higher threshold ( 0.5+). If        recall (capturing all possible subscribers) is more important, use lower thresholds ( 0.2â€“0.3).Based on this, LightGBM and XGBoost are far more stable and reliable than AdaBoost.

Principal Component Analysis (PCA) is a dimensionality reduction technique that transforms high-dimensional data into a lower-dimensional form while preserving most of the variance. 

```python
# Preprocessing the data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', MinMaxScaler())
        ]), ['day',  'campaign', 'total_contact_time', 'contact_intensity']),
        
        ('cat', Pipeline([
            ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
            ('encoder', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'))
        ]), ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month',  'age_group', 'balance_group', 'call_duration_group'])
    ])

x = multi_attempt_converters.drop('y',axis=1)
x_preprocessed = preprocessor.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(x_preprocessed)  
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.axhline(y=0.95, color='r', linestyle='--')  # 95% variance threshold
plt.show()
```

This code performs dimensionality reduction using PCA on top 7 most important features, and the visualization shows how these features distribute data in a simplified 2D space. #### Fit K-Means on PCA reduced data

```python
# Use your PCA-transformed data (x_pca with 5 components)
data_for_clustering = x_pca  # Shape: (n_samples, 5)

# Find optimal K using Elbow Method & Silhouette Score
wcss = []  # Within-Cluster-Sum-of-Squares
silhouette_scores = []
k_range = range(2, 8)  # Test K from 2 to 7

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=seed)
    kmeans.fit(data_for_clustering)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(data_for_clustering, kmeans.labels_))

# Plot Elbow Method
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(k_range, wcss, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.title('Elbow Method')

# Plot Silhouette Scores
plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, 'go-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis')
plt.tight_layout()
plt.show()
```

### Clustering Evaluation: Choosing Optimal K

| Method             | Observation                                                                 |
|--------------------|-----------------------------------------------------------------------------|
| **Elbow Method**   | The 'elbow' is noticeable around **K=4**, indicating diminishing returns in WCSS reduction beyond this point. |
| **Silhouette Score** | Silhouette score increases with more clusters, but improvement flattens beyond **K=5**. |

**Conclusion:**  
Choosing **K=4** provides a balance between compact clusters (low WCSS) and well-separated clusters (high silhouette score). This makes it a suitable choice for interpretable segmentation.

Using 5 PCA components (x_pca) as input for clustering (reducing noise while keeping 96% of variance).

```python
#Choose Optimal K & Final Clustering
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"Optimal K: {optimal_k} (Silhouette Score: {max(silhouette_scores):.2f})")

# Fit final K-Means with optimal K
final_kmeans = KMeans(n_clusters=optimal_k, random_state=seed)
clusters = final_kmeans.fit_predict(data_for_clustering)

# Add clusters back to original data
multi_attempt_converters['Cluster'] = clusters 
```
